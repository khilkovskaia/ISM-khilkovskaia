```python
import pandas as pd
Location = "gradedata-1.csv"
df = pd.read_csv(Location)
df.head()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(
    formula='grade ~ age + exercise + hours + gender_b',
    data=df).fit()
result.summary()
```
